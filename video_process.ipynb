```python
speaker_timestamp = []
# print the result
for turn, _, speaker in diarization.itertracks(yield_label=True):
    print(f"start={turn.start:.1f}s stop={turn.end:.1f}s speaker_{speaker}")
    speaker_ = {"timestamp": [turn.start, turn.end], "speaker": speaker}
    speaker_timestamp.append(speaker_)
```

```python
def merge_speaker_timestamps(timestamps):
    merged_timestamps = []
    for entry in timestamps:
        if merged_timestamps and merged_timestamps[-1]['speaker'] == entry['speaker']:
            # Extend the end time of the last timestamp
            merged_timestamps[-1]['timestamp'][1] = entry['timestamp'][1]
        else:
            # Add a new entry to the merged list
            merged_timestamps.append(entry.copy())
    return merged_timestamps

# Merged timestamps
merged_speaker_timestamps = merge_speaker_timestamps(speaker_timestamp)



def convert_to_milliseconds(timestamps):
    for entry in timestamps:
        # Converting start and end timestamps to milliseconds
        entry['timestamp'][0] = int(entry['timestamp'][0] * 1000)
        entry['timestamp'][1] = int(entry['timestamp'][1] * 1000)
    return timestamps

converted_timestamps = convert_to_milliseconds(merged_speaker_timestamps)
converted_timestamps


```

```python
import torch
from transformers import AutoModelForSpeechSeq2Seq, AutoProcessor, pipeline
from datasets import load_dataset
import sys
sys.path.append('C:\\Users\\a8517\\IdeaProjects\\whisper\\ffmpeg\\bin')


device = "cuda:0" if torch.cuda.is_available() else "cpu"
print(f"device is {device}")
torch_dtype = torch.float16 if torch.cuda.is_available() else torch.float32

model_id = "openai/whisper-large-v3"

model = AutoModelForSpeechSeq2Seq.from_pretrained(
    model_id, torch_dtype=torch_dtype, low_cpu_mem_usage=True, use_safetensors=True
)
model.to(device)

processor = AutoProcessor.from_pretrained(model_id)

pipe = pipeline(
    "automatic-speech-recognition",
    model=model,
    tokenizer=processor.tokenizer,
    feature_extractor=processor.feature_extractor,
    max_new_tokens=128,
    chunk_length_s=30,
    batch_size=16,
    return_timestamps=True,
    torch_dtype=torch_dtype,
    device=device,
)
```

```python
def merge_chunk(chunks):
    texts=[entry['text'] for entry in chunks]
    return ','.join(texts)

```

```python
import os

def get_filenames_in_directory(directory):
    """ 获取指定文件夹下所有文件的名称 """
    filenames = []
    for root, dirs, files in os.walk(directory):
        for file in files:
            filenames.append(file)
    return filenames

# 使用示例
# 替换 'your_directory_path' 为你的目标文件夹路径
directory_path = './segment'
file_names = get_filenames_in_directory(directory_path)
```

```python
for file in file_names:
    file_path = f'''./segment/{file}'''
    result = pipe(file_path, return_timestamps=True)
    msg = merge_chunk(result['chunks'])
    txt_file = file.replace(".wav", ".txt")
    with open(f'./segment_txt/{txt_file}', 'w', encoding='utf-8') as file:
        file.write(msg)
```

```python
dialogues = []
temp_index = 0
for stamp in merged_speaker_timestamps:
    speaker = stamp['speaker']
    file_name = f'./segment_txt/{temp_index}_{speaker}.txt'
    with open(file_name, 'r', encoding='utf-8') as file:
        msg = file.read()
        dialogue = {"speaker": speaker, "content": msg}
        dialogues.append(dialogue)
    temp_index = temp_index + 1    
        
```
